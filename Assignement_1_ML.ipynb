```python
def predict(classLabel):
    class_label_range = classLabel*240
    X = np.zeros((2400, 784))
    t = np.genfromtxt('Train/Training Labels.txt')
    for i in range (0, X.shape[0]):
        img = plt.imread("Train/"+str(i+1)+'.jpg')
        rows, cols = img.shape 
        img_size = rows*cols
        img_1D_vector = img.reshape(img_size)
        X[i] = img_1D_vector
    for j in range(0, t.shape[0]):
        if (t[j] == classLabel):
            t[j] = 1
        else:
            t[j] = 0
    X_trans = np.transpose(X)
    X_X_transpose_inv = np.linalg.pinv(np.dot(X_trans, X))
    eqt = np.dot(np.dot((X_X_transpose_inv), X_trans), t)
    return eqt
predict_0 = predict(0)
predict_1 = predict(1)
predict_2 = predict(2)
predict_3 = predict(3)
predict_4 = predict(4)
predict_5 = predict(5)
predict_6 = predict(6)
predict_7 = predict(7)
predict_8 = predict(8)
predict_9 = predict(9)
```
